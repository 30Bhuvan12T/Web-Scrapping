```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.rolex.com/watches/find-rolex?group=0"
```

```python
url
```

```python
web = requests.get(url)
web
```

```python
web.content
```

```python
soup = BeautifulSoup(web.content,"html.parser")
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.p.string
```

```python
soup.find("p")
```

```python
lines = soup.find_all("p")
for l in lines:
    print(l.text)
```

```python
soup.find_all("h1")
```

```python
product_name = []
product_desc = []
product_price = []
```

```python
name = soup.find_all("h2",class_="css-e76f16-WatchCard-Title-text-body50-text-bold-text-legend100 e1wd4zgb6")
print(name)
```

```python
for i in name:
    product_name.append(name)

print(product_name)
```

```python
desc = soup.find_all("p",class_="css-8nx2zo-WatchCard-Para-text-legend100-text-legend80 e1wd4zgb5")
print(desc)
```

```python
for i in desc:
    product_desc.append(desc)

print(product_desc)
```

```python
price = soup.find_all("span",class_="price")
print(price)
```

```python
price_spans = soup.find_all("span",class_="price")
for span in price_spans:
    product_price.append(span)
    print(span.text.strip())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"product_name":product_name,"product_desc":product_desc,"product_price":product_price})
df
```
